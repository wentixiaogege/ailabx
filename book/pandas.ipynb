```python
df['rate'] = df['close'].pct_change()
df['rate'].plot()
```

```python
df['rate2'] = df['close'] / df['close'].shift(1) -1 # shift往“下”移一位
df
```

```python
#1元波动图
df['equity'] = (df['rate'] +1).cumprod()
#max_equity 为有史以来的最大值序列
df['max_equity'] = df['equity'].expanding().max()
df['max_rolling'] = df['equity'].rolling(200).max()
mdd = round((df['equity'] / df['equity'].expanding(min_periods=1).max()).min() - 1, 3)
print(mdd)
df[['equity','max_equity','max_rolling']].plot()
```

```python
import pandas as pd
amount = pd.Series([100, 90, 110, 150, 110, 130, 80, 90, 100, 150])
print(amount.rolling(3).sum())
print(amount.rolling(3,min_periods=1).sum())
print(amount.rolling(3).mean())
print(amount.rolling(3,min_periods=1).mean())
print("=============max===================")
print(amount.rolling(3).max())
print(amount.rolling(3,min_periods=1).max())

print("=============min===================")
print(amount.rolling(3).min())
print(amount.rolling(3,min_periods=1).min())
```

```python
def load_data(name):
    df = pd.read_csv('../data/csv/{}.csv'.format(name))
    df['date'] = df['date'].apply(lambda x:str(x))
    df.set_index('date',inplace=True)
    df.sort_index(ascending=True,inplace=True)
    df[name] = (1+df['close'].pct_change()).cumprod()
    df = df[[name]]
    #df.rename(columns={'close':name}, inplace=True)
    return df
df_300 = load_data('000300.SH')
df_spx = load_data('SPX')
print(df_300.head(),df_spx.head())
```

```python
df_all = pd.concat([df_300,df_spx], axis=1)
df_all.sort_index(ascending=True, inplace=True)
df_all
```
